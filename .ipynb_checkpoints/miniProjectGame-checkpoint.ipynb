```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
SHEET_URL = 'https://docs.google.com/spreadsheets/d/10poofg-l8DMdtUgGy8mOpra2IHmA9EQC7drmF9AyYHA/export?format=xlsx'
data_src_xls = pd.ExcelFile(SHEET_URL)
data_xls = pd.read_excel(data_src_xls)
```

```python
data_xls.head()
```

```python
data_xls.info()
```

```python
data_cleand = data_xls.copy()
```

```python
duplicated_name = data_cleand[data_cleand['Name'].duplicated() == True]
duplicated_name
```

```python
data_cleand_without_duplicated_name = data_cleand[data_cleand['Name'].duplicated() == False]
```

```python
data_cleand_without_duplicated_name.info()
```

```python
cleaned_data = data_cleand_without_duplicated_name.copy()
cleaned_data.head()
```

```python
game_sort_asc = cleaned_data.sort_values(by='Release', ascending=True, ignore_index=True).head(3)[['Name', 'Release']]
game_sort_desc = cleaned_data.sort_values(by='Release', ascending=True, ignore_index=True).tail(3)[['Name', 'Release']]
game_sort = pd.concat([game_sort_asc, game_sort_desc], ignore_index=True)
```

```python
game_sort
```

```python
plt.figure(figsize=(12,5))
plt.bar(game_sort['Name'], game_sort['Release'])
plt.xlabel('Game Name')
plt.ylabel('Year')
plt.title('The Top 3 and Bottom 3 of The Game')
plt.show()
```
